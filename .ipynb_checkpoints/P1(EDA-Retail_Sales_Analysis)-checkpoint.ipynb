```python
#OUTLIERS:
import matplotlib.pyplot as plt
import seaborn as sns
# Create histograms for numeric columns
numeric_columns = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
for i,col in enumerate(numeric_columns):
    sns.histplot(df[col], ax=axes[i//2, i%2], kde=True)
    axes[i//2, i%2].set_title(f'Histogram of {col}')
    axes[i//2, i%2].set_xlabel(col)
    axes[i//2, i%2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create boxplots for numeric columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, col in enumerate(numeric_columns):
    sns.boxplot(x=df[col], ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'Boxplot of {col}')
    axes[i//2, i%2].set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Create count plots for categorical columns
categorical_columns = ['Gender', 'Product Category']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

for i, col in enumerate(categorical_columns):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#Load the data from a csv file.
df=pd.read_csv("C:\\Users\\khush\\Desktop\\Data Analysts\\OASIS INFOBYTE\\Retail_Sales_Datasets\\retail_sales_dataset.csv")
```

```python
print(df.head())
#This helps us understand the structure and content of the data.
```

```python
# Extracting year from the date column
df['Year'] = pd.to_datetime(df['Date']).dt.year

plt.figure(figsize=(10, 10))
plt.plot(df['Year'], df['Age'], label='Age')
plt.xlabel('Year')
plt.ylabel('Age group of customer')
plt.title('Yearly sold products on different Age group')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm

# Assuming 'Date' column contains datetime values
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Perform seasonal decomposition
decomposition = sm.tsa.seasonal_decompose(df['Age'], model='additive', period=12)  # Assuming yearly seasonality (period=12)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(df['Age'], label='Original', color='blue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend', color='red')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonal', color='green')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residual', color='purple')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
result = sm.tsa.adfuller(df['Age'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```
