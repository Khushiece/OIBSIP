# Data Cleaning

Data cleaning, also known as data cleansing or data scrubbing, is the process of identifying and correcting errors, inconsistencies, and inaccuracies in datasets to improve their quality and reliability for analysis or other purposes. 

Here are some common tasks involved in data cleaning:

1.Handling missing values: Identify and deal with missing values in the dataset, which can include imputation (replacing missing values with estimated or calculated values), deletion of rows or columns with missing values, or marking missing values for special treatment during analysis.

2.Removing duplicate records: Identify and remove duplicate entries from the dataset to avoid redundancy and ensure data integrity.

3.Standardizing data formats: Ensure consistency in data formats across the dataset, such as date formats, numerical representations, and text encoding.

4.Correcting errors: Identify and correct errors in the data, such as typographical errors, incorrect values, or inconsistencies in coding schemes.

5.Handling outliers: Detect and address outliers or anomalies in the data that may skew analysis results or model performance.

6.Normalizing data: Scale and transform data to bring it into a consistent format or range, which can improve the effectiveness of analysis and modeling techniques.

# Missing Values

```python
#Looking for missing 
data.isna().sum()
```

```python
missing_values=data.isnull()# Check for missing values
missing_values
```

False indicates that the corresponding element is not missing, while True indicates that it is missing.

```python
# Get the indices where missing_values is True
missing_indices = missing_values[missing_values].stack().index
```

```python
# Iterate through the indices and print the corresponding values
for row, col in missing_indices:
    print(f"Missing value at row {row}, column '{col}': {data.at[row, col]}")
```

# Dealing or Resolving the missing values.


Fill missing values with a specific constant, mean, median, or mode using the fillna() method.

Since Column_name-name and host name are categorical data and having Dtypes Object and number of missing values equal to 16 and 21 respectively..

```python
# Fill missing values in 'name' with the most frequent name
data['name'].fillna(data['name'].mode()[0], inplace=True)

# Fill missing values in 'host_name' with a special category like "Unknown"
data['host_name'].fillna('Unknown', inplace=True)
```

For the reviews_per_months column representing the average number of reviews per month (float data type)

```python
# Fill missing values in 'reviews_per_months' with the mean value
data['reviews_per_month'].fillna(data['reviews_per_month'].median(), inplace=True)
```

For the last_review column representing dates,we resolve the missing values problems by fill missing values with mode(represents the maximum frequency of data)

```python
data['last_review'].fillna(data['last_review'].mode()[0], inplace=True)
```

```python
data.isna().sum()
```

Now all the missing values are removed..

# Outlier Detection: 

Identifying and addressing outliers that may skew analysis or model
performance.


Outliers refer to data points that significantly differ from other observations in a dataset. These data points can skew statistical analyses and models, potentially leading to incorrect conclusions if not properly addressed. Identifying and handling outliers is an important step in data preprocessing and analysis

```python
# Visualize distributions
numeric_columns = ['id', 'host_id', 'latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

for col in numeric_columns:
    plt.figure(figsize=(8, 6))
    data[col].hist(bins=30)
    plt.title(col)
    plt.show()
```

```python
# Define outlier criteria (e.g., using z-scores)
threshold = 3
outlier_criteria = (data[numeric_columns] - data[numeric_columns].mean()).abs() > threshold * data[numeric_columns].std()
```

```python
# Identify outliers
outliers = data[outlier_criteria.any(axis=1)]
```

# Handling Outliers.

```python
cleaned_data = data[~outlier_criteria.any(axis=1)]
```

```python
plt.figure(figsize=(6, 4))  
sns.boxplot(data=cleaned_data, y='price', palette='viridis')
plt.title('Box Plot for price')
plt.ylabel('price')
plt.show()
```
